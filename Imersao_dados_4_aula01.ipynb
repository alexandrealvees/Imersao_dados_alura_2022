1 - Realizar a média da metragem para cada um dos bairros (Paulo)

```python
import numpy as np

mediaMetragemBairro = np.around(dados.groupby(['Bairro']).agg(Media=('Metragem', 'mean')), decimals=2)
mediaMetragemBairro.head(10).sort_values(by=('Media'), ascending = False)
```

3 - Explorar alguns gráficos na documentação e aplicar nas demais colunas do DF, tentar colocar alguma conclusão (Thiago)

```python
from numpy.lib.shape_base import column_stack
metragemMenor100 = dados.query('Metragem < 100')
metragemDe100a300 = dados.query('Metragem >= 100 & Metragem <= 300')
metragemDe301a500 = dados.query('Metragem >= 301 & Metragem <= 500')
metragemMaior500 = dados.query('Metragem >= 501')

countFaixa1 =  metragemMenor100["Cidade"].count()
countFaixa2 =  metragemDe100a300["Cidade"].count()
countFaixa3 =  metragemDe301a500["Cidade"].count()
countFaixa4 =  metragemMaior500["Cidade"].count()

#contagemPorFaixa = pd.DataFrame({'-100': [countFaixa1], '101-300': [countFaixa2], '301-500': [countFaixa3], '500+': [countFaixa4]})

contagemPorFaixa = pd.DataFrame({'Faixas de metragem': [countFaixa1, countFaixa2, countFaixa3, countFaixa4]}, 
                                index=['100-', '101-300', '301-500', '500+'],
                                )


contagemPorFaixa.plot.pie(y='Faixas de metragem', colors=["r", "g", "b", "c"],
    autopct="%.2f",
    fontsize=20,
    figsize=(6, 6))
```
